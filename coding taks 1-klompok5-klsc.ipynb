Import Dataset dan Library

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data = pd.read_excel("Dataset.xlsx")
data
```

Menampilkan Dataset

```python
# Menghitung statistik
hujan_mean = data['Curah Hujan (mm)'].mean()
mean_a = data['Penjualan A (pcs)'].mean()
mean_b = data['Penjualan B (pcs)'].mean()

hujan_median = data['Curah Hujan (mm)'].median()
median_a = data['Penjualan A (pcs)'].median()
median_b = data['Penjualan B (pcs)'].median()

hujan_modus = data['Curah Hujan (mm)'].mode().iloc[0]
modus_a = data['Penjualan A (pcs)'].mode().iloc[0]
modus_b = data['Penjualan B (pcs)'].mode().iloc[0]

hujan_range = data['Curah Hujan (mm)'].max() - data['Curah Hujan (mm)'].min()
range_a = data['Penjualan A (pcs)'].max() - data['Penjualan A (pcs)'].min()
range_b = data['Penjualan B (pcs)'].max() - data['Penjualan B (pcs)'].min()

IQR_hujan = stats.iqr(data['Curah Hujan (mm)'])
IQR_a = stats.iqr(data['Penjualan A (pcs)'])
IQR_b = stats.iqr(data['Penjualan B (pcs)'])

varian_hujan = data['Curah Hujan (mm)'].var()
varian_a = data['Penjualan A (pcs)'].var()
varian_b = data['Penjualan B (pcs)'].var()

hujan_skew = data['Curah Hujan (mm)'].skew()
skew_a = data['Penjualan A (pcs)'].skew()
skew_b = data['Penjualan B (pcs)'].skew()

hujan_kurtosis = data['Curah Hujan (mm)'].kurtosis()
kurtosis_a = data['Penjualan A (pcs)'].kurtosis()
kurtosis_b = data['Penjualan B (pcs)'].kurtosis()

# Membuat DataFrame langsung tanpa dictionary
df_tampil = pd.DataFrame([
    ['Mean', hujan_mean, mean_a, mean_b],
    ['Median', hujan_median, median_a, median_b],
    ['Modus', hujan_modus, modus_a, modus_b],
    ['Range', hujan_range, range_a, range_b],
    ['IQR', IQR_hujan, IQR_a, IQR_b],
    ['Varians', varian_hujan, varian_a, varian_b],
    ['Skew', hujan_skew, skew_a, skew_b],
    ['Kurtosis', hujan_kurtosis, kurtosis_a, kurtosis_b]
], columns=['Statistik', 'Curah Hujan (mm)', 'Penjualan A (pcs)', 'Penjualan B (pcs)'])

# Menampilkan hasil
print(df_tampil)
```

```python
def detect_outliers(column):
    Q1 = np.percentile(column, 25)
    Q3 = np.percentile(column, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = column[(column < lower_bound) | (column > upper_bound)]
    return outliers

# Mendeteksi outlier untuk setiap variabel
outliers_hujan = detect_outliers(data['Curah Hujan (mm)'])
outliers_a = detect_outliers(data['Penjualan A (pcs)'])
outliers_b = detect_outliers(data['Penjualan B (pcs)'])

# Menghapus outlier dari dataset
data_cleaned = data[~data['Curah Hujan (mm)'].isin(outliers_hujan) &
                     ~data['Penjualan A (pcs)'].isin(outliers_a) &
                     ~data['Penjualan B (pcs)'].isin(outliers_b)]

# Menampilkan boxplot untuk data yang sudah dibersihkan
plt.figure(figsize=(10, 15))

# Boxplot untuk 'Curah Hujan (mm)'
plt.subplot(3, 1, 1)
plt.boxplot(data_cleaned['Curah Hujan (mm)'], vert=True)  # vert=True untuk boxplot vertikal
plt.title('Boxplot Curah Hujan (mm) - Data Bersih')
plt.ylabel('Curah Hujan (mm)')

# Boxplot untuk 'Penjualan A (pcs)'
plt.subplot(3, 1, 2)
plt.boxplot(data_cleaned['Penjualan A (pcs)'], vert=True)  # vert=True untuk boxplot vertikal
plt.title('Boxplot Penjualan A (pcs) - Data Bersih')
plt.ylabel('Penjualan A (pcs)')

# Boxplot untuk 'Penjualan B (pcs)'
plt.subplot(3, 1, 3)
plt.boxplot(data_cleaned['Penjualan B (pcs)'], vert=True)  # vert=True untuk boxplot vertikal
plt.title('Boxplot Penjualan B (pcs) - Data Bersih')
plt.ylabel('Penjualan B (pcs)')

plt.tight_layout()  # Mengatur jarak antar plot
plt.show()
```
